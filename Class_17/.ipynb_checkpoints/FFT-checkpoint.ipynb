```python
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
# Creamos una señal de ejemplo
t = np.linspace(0, 1, 1000)
f1 = 10
f2 = 50
y = np.sin(2*np.pi*f1*t) + 2*np.cos(2*np.pi*f2*t)

# Calculamos la FFT
fft_y = np.fft.fft(y)

# Calculamos las frecuencias correspondientes a cada componente de la FFT
freqs = np.fft.fftfreq(len(y), t[1]-t[0])


plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
plt.plot(t, y, color='#228B22')
plt.title('y(n)', fontsize=18)
plt.xlabel('Tiempo (s)', fontsize=16)
plt.ylabel('Amplitud', fontsize=16)
plt.grid()

plt.subplot(2,1,2)
plt.plot(freqs, np.abs(fft_y), color='#20B2AA')
plt.title('Espectro de y(n)', fontsize=18)
plt.xlabel('Frecuencia (Hz)', fontsize=16)
plt.ylabel('Amplitud', fontsize=16)
plt.xlim(-100,100)
plt.grid()

plt.tight_layout()
```

- En el espectro de la señal se observan las dos frecuencias que la componen, 10 Hz y 50 Hz.
- Aparecen 4 picos dado que el espectro es simétrico.
- La amplitud de los picos depende de la amplitud que tiene las señales en el tiempo.
