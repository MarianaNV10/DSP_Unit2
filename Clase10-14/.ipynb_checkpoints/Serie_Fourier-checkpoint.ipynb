```python
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
N = 16 #Número de muestras
n = np.arange(-int(N/2), int(N/2)) #Vector de tiempo
x_n1 = []
x_n2 = np.zeros(int(N/2))

for i in range(0,int(N/2)):
    x_n1.append(2*i)

cont = 1
for j in range(int(N/2)-1,0,-1):
    x_n2[j] = x_n1[cont]
    cont += 1

print(x_n1) 
print(x_n2)

x_n = np.hstack((x_n2, x_n1))

plt.stem(n,x_n, linefmt='#5F9EA0')
plt.title('x(n)', fontsize=18)
plt.xlabel('Sample', fontsize=16)
plt.ylabel('Amplitude', fontsize=16)
plt.grid()
```

```python
K = 10 #Coeficientes a calcular
range_k = np.arange(0,K+1)

re_ck = np.zeros(K+1)
im_ck = np.zeros(K+1)
Ck = np.empty(K+1,'complex_') #Coeficientes con parte real e imaginaria

for i in range(0, K+1):
    re_ck[i] = (1/N)*sum(x_n*np.cos((2*np.pi*i*n)/N))
    im_ck[i] = (1/N)*sum(x_n*np.sin((2*np.pi*i*n)/N))
    Ck[i] = (1/N)*sum((x_n)*np.exp((-2j*np.pi*i*n)/N))
```

```python
par = re_ck**2
impar = im_ck**2

plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.stem(range_k, par)
plt.title('Espectro de Potencia de la señal Par', fontsize=18)
plt.xlabel('K', fontsize=16)
plt.ylabel('Amplitude', fontsize=16)
plt.grid()
plt.subplot(2,1,2)
plt.stem(range_k, impar)
plt.title('Espectro de Potencia de la señal Impar', fontsize=18)
plt.xlabel('K', fontsize=16)
plt.ylabel('Amplitude', fontsize=16)
plt.grid()

plt.tight_layout()
```

- Se tiene que x(n) es una señal par, por lo que, los coeficientes que tienen un mayor aporte al espectro en potencia son los pares y esto se puede evidenciar en las gráficas anteriores.

```python
n_ck = np.arange(0,len(Ck),1)
n_ck2 = np.arange(-len(Ck)+1,1)
pot_coef = abs(Ck)**2

plt.figure(figsize=(8,4))
plt.stem(n_ck,pot_coef)
plt.stem(n_ck2,np.flip(pot_coef))
plt.title('Espectro de potencia', fontsize=18)
plt.xlabel('Ck', fontsize=16)
plt.ylabel('|Ck|^2')
plt.grid()
```

- Se observar que, respecto al espectro en potencia, esta señal tiene un nivel DC por el hecho de ser par.
